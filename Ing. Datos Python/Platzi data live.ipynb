<h2> Solicitudes a la Web </h2>

```python
import requests

response = requests.get('https://www.platzi.com')

```

<h3> Extraccion de informacion del HTML </h3>

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')
#print(soup.title.text)

#print(soup.select('meta[name=description]')[0]['content'])
courses_links = soup.select('.CarrersList')
courses = [course['href'] for course in courses_links]

for course in courses:
    print(course)
```
